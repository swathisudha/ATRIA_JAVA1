```python
from docx import Document
import PyPDF2

def read_resume(file_path):
    text = ""
    if file_path.endswith(".docx"):
        doc = Document(file_path)
        for para in doc.paragraphs:
            text += para.text + " "
    elif file_path.endswith(".pdf"):
        with open(file_path, "rb") as f:
            reader = PyPDF2.PdfReader(f)
            for page in reader.pages:
                text += page.extract_text()
    return text
```

```python
resume_text = read_resume("resume.pdf")
print(resume_text[:500])
```

```python
with open("jd.txt", "r", encoding="latin-1") as file:
    jd_text = file.read()

print(jd_text)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

skills_list = [
    "python", "machine learning", "deep learning",
    "nlp", "fastapi", "sql", "data analysis",
    "scikit-learn", "tensorflow", "pandas"
]

def extract_skills(text):
    text = text.lower()
    found_skills = [skill for skill in skills_list if skill in text]
    return list(set(found_skills))
```

```python
resume_skills = extract_skills(resume_text)
jd_skills = extract_skills(jd_text)

print("Resume Skills:", resume_skills)
print("JD Skills:", jd_skills)
```

```python
def match_score(resume_skills, jd_skills):
    if not jd_skills:
        return 0
    matched = set(resume_skills).intersection(set(jd_skills))
    score = (len(matched) / len(jd_skills)) * 100
    return round(score, 2), matched
```

```python
score, matched_skills = match_score(resume_skills, jd_skills)
print("Match Score:", score)
print("Matched Skills:", matched_skills)
```

```python
THRESHOLD = 60  # You can say this in interview

if score >= THRESHOLD:
    print("Candidate shortlisted for next round")
else:
    print("Candidate not shortlisted")
```

```python
def generate_questions(skills):
    questions = []
    for skill in skills:
        questions.append(f"Explain your experience with {skill}.")
    return questions
```

```python
if score >= THRESHOLD:
    questions = generate_questions(matched_skills)
    for q in questions:
        print(q)
```
